**NUMPY BASICS** 

**Single-dimensional Numpy Array:**

**Python NumPy Array v/s List**

```python
import time
import sys
S= range(1000)
print(sys.getsizeof(5)*len(S))
 
D= np.arange(1000)
print(D.size*D.itemsize)
```

```python
SIZE = 1000000
 
L1= range(SIZE)
L2= range(SIZE)
A1= np.arange(SIZE)
A2=np.arange(SIZE)
 
start= time.time()
result=[(x,y) for x,y in zip(L1,L2)]
print((time.time()-start)*1000)
 
start=time.time()
result= A1+A2
print((time.time()-start)*1000)
```

**Python NumPy Operations**

ndim:

```python
a = np.array([(1,2,3),(4,5,6)])
print(a.ndim)
```

itemsize:

```python
a = np.array([(1,2,3)])
print(a.itemsize)
```

dtype:

```python
a = np.array([(1,2,3)])
print(a.dtype)
```

```python
a = np.array([(1,2,3,4,5,6)])
print(a.size)
print(a.shape)
```

reshape:

```python
a = np.array([(8,9,10),(11,12,13)])
print(a)
a=a.reshape(3,2)
print(a)
```

slicing:

```python
a=np.array([(1,2,3,4),(3,4,5,6)])
print(a[0,2])
```

```python
a=np.array([(1,2,3,4),(3,4,5,6)])
print(a[0:,2])
```

```python
a=np.array([(8,9),(10,11),(12,13)])
print(a[0:2,1])
```

linspace

```python
a=np.linspace(1,3,10)
print(a)
```

max/ min

```python
a= np.array([1,2,3])
print(a.min())
print(a.max())
print(a.sum())
```

```python
a= np.array([(1,2,3),(3,4,5)])
print(a.sum(axis=0))
```

Square Root & Standard Deviation

```python
a=np.array([(1,2,3),(3,4,5,)])
print(np.sqrt(a))
print(np.std(a))
```

Addition Operation

```python
x= np.array([(1,2,3),(3,4,5)])
y= np.array([(1,2,3),(3,4,5)])
a=x+y
s=x-y
m=x*y
d=x/y
print(a)
print(s)
print(m)
print(d)
```

Vertical & Horizontal Stacking

```python
x= np.array([(1,2,3),(3,4,5)])
y= np.array([(1,2,3),(3,4,5)])
print(np.vstack((x,y)))
print(np.hstack((x,y)))
```

ravel

```python
x= np.array([(1,2,3),(3,4,5)])
print(x.ravel())
```

**Python Numpy Special Functions**

```python
import matplotlib.pyplot as plt
x= np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
y=np.tan(x)
plt.plot(x,y)
plt.show()
```

```python
a= np.array([1,2,3])
print(np.exp(a))
```

```python
a= np.array([1,2,3])
print(np.log(a))
```

```python
a= np.array([1,2,3])
print(np.log10(a))
```
